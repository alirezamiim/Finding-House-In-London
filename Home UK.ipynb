```python
import csv
from  selenium import webdriver 
import pandas as pd
from bs4 import BeautifulSoup
import re
import threading
```

```python
def ReplacingSpaceWithDash(_str=''):
    return _str.replace(' ','-')
def LowerCase(_str=''):
    return _str.lower()
```

```python
data = pd.read_csv('Stations.csv').drop(columns=['Unnamed: 3'	,'Unnamed: 4',	'Unnamed: 5',	'Unnamed: 6','Line'])
data= data[data['Ok']==True]
```

```python
data.head()
```

```python
data['Station Name'] = data['Station Name'].apply(ReplacingSpaceWithDash)
data['Station Name'] = data['Station Name'].apply(LowerCase)
```

```python
data.tail()
```

```python
for house in range(len(data)):
    station = data.iloc[house]['Station Name']
    print(station,'started')
    temp_url = 'https://www.onthemarket.com/to-rent/property/'+\
        station+'-station/?max-bedrooms=1&max-price=1500&min-price=700&radius=0.5&view=grid'

    driver = webdriver.Chrome()
    driver.get(temp_url)
    text = driver.page_source
    # soup = BeautifulSoup(text)
    driver.close()

    # houses = soup.find_all(attrs={'class':'grid-list-tabcontent hasSpotlight'})
    # print(houses)
    # houses = houses[0]
    houses = str(houses)
    houses = re.findall(r'result-(\d\d+)',text)
    houses.insert(0,station)
    csv_file = station+".csv"


    with open(csv_file, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(houses)
    print('Area',data.iloc[house]['Station Name'], 'hase been done!',f'[{house*100/len(data)}]')
```
