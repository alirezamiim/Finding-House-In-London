```python
for stations in range(len(data)):
    station = data.iloc[stations]['Station Name']
    print(station,'started')
    file = open('Stations/'+'acton-town'+'.csv',mode='r')
    house_ref_no = file.read().split(',')[1:]
    file.close()
    house_list=[]
    for house_no in house_ref_no:
        text = ''
        temp_url= ''
        if len(house_no)>4:
            temp_url = 'https://www.onthemarket.com/details/'+house_no+'/'
            # print(temp_url)
            driver = webdriver.Chrome()
            driver.get(temp_url)
            text = driver.page_source
            driver.close()
            price = re.findall(r'£(.*?) pcm',text)[0]
            tel = re.findall(r'href=\"tel:(.*?)\" class',text)[0]
            csv_file = 'Houses/'+station+'.csv'
            with open(csv_file, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([temp_url,price,tel])
            



    print('Area',data.iloc[stations]['Station Name'], 'hase been done!',f'[{(stations+1)*100/len(data)}]')
```
